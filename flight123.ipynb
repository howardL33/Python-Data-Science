```python
features = ["DAY_OF_WEEK","SCHEDULED_DEPARTURE","AIR_TIME"]
```

```python
g = sns.PairGrid(flights[features+["over_20_min_late"]], hue="over_20_min_late")
g = g.map_diag(plt.hist, histtype="step", linewidth=3)
g = g.map_offdiag(plt.scatter)
g = g.add_legend()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
# choose some features
flights_clean = flights[features + ["over_20_min_late"]].dropna(how="any")
x_vars = flights_clean[features]
target = flights_clean["over_20_min_late"]

clf = LogisticRegression(random_state=0, solver='lbfgs', multi_class='multinomial', max_iter=500).fit(x_vars, target)

# how did we do?
print(clf.score(x_vars, target))
```

```python
# Confusion Matrix
from sklearn.metrics import confusion_matrix

# make predictions
iris_predictions = clf.predict(x_vars)

c_matrix = confusion_matrix(iris_predictions, target)
print(c_matrix)
```
